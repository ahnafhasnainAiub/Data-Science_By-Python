In the first cell, we are simply creating data x and y to use in the provided tasks.

```python
# We need data to work with, lets create data first!
import numpy as np
x = np.arange(0,10)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
# plt.show() for non-notebook users
```

**1. Instructions: **
* ** Use <code>plt.figure()</code> and create a figure object, say fig.**
* ** Add an axis to the figure canvas (`fig`), `[left, bottom, width, height] is [0,0,1,1]`. <i>Hint: <code>add_axes()</code><i>**
* ** Plot the data x,y on the axes**
* ** Set the labels x and y**
* ** Set the title**
* ** Create legends as shown in the plot**

Match the output below: 

```python
import matplotlib.pyplot as plt
fig1 = plt.figure(figsize=(4,3))
axes = fig1.add_axes([0,0,1,1]) 

axes.plot(x, y,'g-')


axes.set_xlabel('x_Label') 
axes.set_ylabel('y_Label')
axes.set_title('Plot_title')

axes.legend(['y = x*2'],loc ="lower right")
```

**2. Instructions**<br>
* ** Create a figure object**
* ** Put two axes on the figure at [0,0,1,1] and [0.1,0.6,0.2,0.2]**
* ** Plot data (x,z) on main axes**
* ** Plot data (x,z) on the small axes and set its x limit b/w 1, 2, y limit 0, 5**
* ** Replicate the figure to axes, titles, legends etc**

```python
fig3 = plt.figure(figsize=(5.5,4))

# Creating main axes 
axes_main = fig3.add_axes([0,0,1,1] )

# # Creating inset axes - values b/w 0 & 1
axes_inset = fig3.add_axes([0.1,0.6,0.2,0.2]) 
axes_inset.set_xlim([1, 2])
axes_inset.set_ylim([0, 5])

# # Main figure
axes_main.plot(x, z, 'r-')
axes_main.set_xlabel('X_main')
axes_main.set_ylabel('Y_main')
axes_main.set_title('Title_main')

# # Inset figure
axes_inset.plot(x, z, 'b')
axes_inset.set_xlabel('x_inset')
axes_inset.set_ylabel('y_inset')
axes_inset.set_title('Inset');

axes_main.legend(["y_main = x_main**2"], loc ="lower right")
```

**3. Replicate the output using <code>subplots()</code> and for loop** -- Note, the data is x and y. 

```python
fig4, axes1 = plt.subplots(nrows=1, ncols=3,figsize=(5,3)) 
for ax in axes1:
    ax.plot(x, y)
    ax.set_title('Title')
fig4.tight_layout() 
```
